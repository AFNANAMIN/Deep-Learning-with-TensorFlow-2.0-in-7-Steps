### Split into Training and Test

```
df_train = df.loc[:299, :]
df_test = df.loc[299:, :]
df_train.info()
```

### Split into features and labels

```
feature_names = ['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']
target_name = 'medv'
```

```
features_train = df_train[feature_names]
target_train = df_train[target_name]

features_test = df_test[feature_names]
target_test = df_test[target_name]
```

```
features_train = features_train.as_matrix()
features_test = features_test.as_matrix()
```

### Build a Linear Regression Model

```
model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(len(feature_names), ))
])
```

```
model.compile(optimizer='adam', loss='mse')
model.fit(features_train, target_train, epochs=500)
```

```
print(model.summary())
```

```
predictions = model.predict(features_test)
print(predictions[0])
```
